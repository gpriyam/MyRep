```python
import pandas as pd
import numpy as np
```

```python
#Read the data into a dataframe
df=pd.read_excel('Toronto_data.xlsx')
```

```python
df=pd.DataFrame(df)
df
```

```python
#Remove all rows where Borough data is not assigned
df_new = df[df.Borough != 'Not assigned']
```

```python
#Check the no. of rows remaining
df_new.info()
```

```python
#Check the sum of frequency of the Postal codes (in this case the data shows all rows are unique and has no repeatation)
df_new['Postal Code'].value_counts().sum()
  
 
```

```python
df_new.reset_index(inplace = True)
df_new
```

```python
#To check the no of remaining rows after Data Wrangling
df_new.shape
```

```python
geo=pd.read_csv('Geospatial_coordinates.csv')
```

```python
geo
```

```python
#Add latitude and longitude column to the dataframe from the geo dataframe after matching the postal code

for i in range(len(df_new)):
    for j in range(len(geo)):
        if (df_new['Postal Code'][i]==geo['Postal Code'][j]):
            df_new['Latitude'][i]=geo['Latitude'][j]
            df_new['Longitude'][i]=geo['Longitude'][j]
            
            
```

```python
#Drop the index column
df_new.drop(axis=1,labels='index')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Make a new dataframe with Boroughs having 'Toronto' in its name
df_tor=df_new[df_new['Borough'].str.contains('Toronto', na = False)]
```

```python
df_tor.reset_index(inplace = True)
```

```python
df_tor.shape
```

```python
df_tor.drop(axis=1,labels=['index','level_0'])
```

```python
#Plottling all Lat longs for all neighbourhoods in Toronto
sns.scatterplot(x=df_tor.Longitude,y=df_tor.Latitude,data=df_tor,hue=df_tor.Borough)
plt.title("Map of Toronto")
plt.show()
```

```python
import folium
```

```python
from geopy.geocoders import Nominatim
```

```python
#Find geographical coordinates of Toronto
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latt = location.latitude
long = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# Plot all Toronto Boroughs on the map
map_toronto = folium.Map(location=[latt, long], zoom_start=12)

for lat, lng, label in zip(df_tor['Latitude'], df_tor['Longitude'], df_tor['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)

# add markers to map

    
map_toronto
```

```python
#KMeans clustering of all neighbourhoods in Toronto based on latitude and longitude
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=4)
km
```

```python
y_pred=km.fit_predict(df_tor[['Latitude','Longitude']])
```

```python
y_pred
```

```python
df_tor['Predicted']=y_pred
```

```python
#Dropping extra columns
df_tor.drop(axis=1,labels=['level_0','index'])
```

```python
#Encoding Boroughs of Toronto into a new column 'Actual'
df_tor['Actual']=0
for i in range(len(df_tor)):
    if(df_tor['Borough'][i]=='Downtown Toronto'):
        df_tor['Actual'][i]=0
    elif(df_tor['Borough'][i]=='East Toronto'):
        df_tor['Actual'][i]=1
    elif(df_tor['Borough'][i]=='Central Toronto'):
        df_tor['Actual'][i]=2
    elif (df_tor['Borough'][i]=='West Toronto'):
        df_tor['Actual'][i]=3
                       
  
```

```python
df_tor
```

```python
#Checking how good were the prediction of Borough
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(df_tor['Actual'],df_tor['Predicted'])#Confusion matrix that the clustering was correct in 37 out of 39 cases
```

```python
accuracy_score(df_tor['Actual'],df_tor['Predicted'])#Confusion matrix that the clustering was correct in 94% cases
```
